```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# This line makes our plots appear in the notebook
%matplotlib inline

print("Libraries imported successfully!")
```

```python
# Step 2: Load the dataset
file_path = r"C:\Users\User\Desktop\ivy\tested.csv"
df = pd.read_csv(file_path)

print("Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")
```

```python
# Step 3: Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())

print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Step 4: Explore dataset structure
print("Dataset Information:")
print(df.info())

print("\nDataset shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\nData types:")
print(df.dtypes)
```

```python
# Step 5: Check for missing values in detail
print("Missing values count:")
print(df.isnull().sum())

print("\nMissing values percentage:")
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent.round(2))
```

```python
# Step 6: Clean the dataset
print("Before cleaning:")
print(f"Shape: {df.shape}")

# Fill missing Age values with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Fare with the median fare
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# For Cabin, we'll create a new category for missing values
df['Cabin'].fillna('Unknown', inplace=True)

print("\nAfter cleaning:")
print("Missing values:")
print(df.isnull().sum())
```

```python
# Step 7: Basic statistics for numerical columns
print("Descriptive Statistics:")
print(df.describe())

print("\nSurvival Rate:")
survival_rate = df['Survived'].mean()
print(f"Overall survival rate: {survival_rate:.2%}")
```

```python
# Step 8: Group analysis by categorical variables
print("Survival Rate by Gender:")
gender_survival = df.groupby('Sex')['Survived'].agg(['count', 'mean'])
gender_survival.columns = ['Total_Passengers', 'Survival_Rate']
print(gender_survival)

print("\nSurvival Rate by Passenger Class:")
class_survival = df.groupby('Pclass')['Survived'].agg(['count', 'mean'])
class_survival.columns = ['Total_Passengers', 'Survival_Rate']
print(class_survival)

print("\nAverage Age by Passenger Class:")
age_by_class = df.groupby('Pclass')['Age'].mean()
print(age_by_class.round(1))
```

```python
# Step 9: Create basic visualizations
plt.figure(figsize=(15, 10))

# 1. Survival Rate by Gender
plt.subplot(2, 3, 1)
gender_counts = df.groupby(['Sex', 'Survived']).size().unstack()
gender_counts.plot(kind='bar', ax=plt.gca())
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(['Died', 'Survived'])
plt.xticks(rotation=0)

# 2. Survival Rate by Class
plt.subplot(2, 3, 2)
class_counts = df.groupby(['Pclass', 'Survived']).size().unstack()
class_counts.plot(kind='bar', ax=plt.gca())
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(['Died', 'Survived'])

# 3. Age Distribution
plt.subplot(2, 3, 3)
plt.hist(df['Age'], bins=20, alpha=0.7, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# 4. Fare Distribution by Class
plt.subplot(2, 3, 4)
for pclass in [1, 2, 3]:
    class_fares = df[df['Pclass'] == pclass]['Fare']
    plt.hist(class_fares, alpha=0.6, label=f'Class {pclass}', bins=20)
plt.title('Fare Distribution by Class')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend()

# 5. Survival Rate by Age Groups
plt.subplot(2, 3, 5)
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 18, 35, 60, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
age_group_survival = df.groupby('Age_Group')['Survived'].mean()
age_group_survival.plot(kind='bar', color='coral')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)

# 6. Embarked Port Analysis
plt.subplot(2, 3, 6)
embarked_survival = df.groupby('Embarked')['Survived'].mean()
embarked_survival.plot(kind='bar', color='lightgreen')
plt.title('Survival Rate by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Survival Rate')

plt.tight_layout()
plt.show()
```

```python
# Step 10: Summary of findings
print("=== KEY FINDINGS FROM TITANIC DATA ANALYSIS ===")
print()
print(f"1. Dataset Overview:")
print(f"   - Total passengers: {len(df)}")
print(f"   - Overall survival rate: {df['Survived'].mean():.1%}")
print()

print("2. Gender Analysis:")
gender_stats = df.groupby('Sex')['Survived'].agg(['count', 'mean'])
for gender in gender_stats.index:
    count = gender_stats.loc[gender, 'count']
    rate = gender_stats.loc[gender, 'mean']
    print(f"   - {gender.capitalize()}: {rate:.1%} survival rate ({count} passengers)")
print()

print("3. Class Analysis:")
class_stats = df.groupby('Pclass')['Survived'].agg(['count', 'mean'])
for pclass in class_stats.index:
    count = class_stats.loc[pclass, 'count']
    rate = class_stats.loc[pclass, 'mean']
    print(f"   - Class {pclass}: {rate:.1%} survival rate ({count} passengers)")
print()

print("4. Age Analysis:")
age_group_stats = df.groupby('Age_Group')['Survived'].agg(['count', 'mean'])
for age_group in age_group_stats.index:
    count = age_group_stats.loc[age_group, 'count']
    rate = age_group_stats.loc[age_group, 'mean']
    print(f"   - {age_group}: {rate:.1%} survival rate ({count} passengers)")
```
